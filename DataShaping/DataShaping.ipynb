```python
import pandas as pd

# Data Source
# https://datos.gob.ar/dataset/energia-produccion-petroleo-gas-por-pozo-capitulo-iv

# List of CSV files to read
csv_files = [
    'produccin-de-pozos-de-gas-y-petrleo-2017.csv',
    'produccin-de-pozos-de-gas-y-petrleo-2018.csv',
    'produccin-de-pozos-de-gas-y-petrleo-2019.csv',
    'produccin-de-pozos-de-gas-y-petrleo-2020.csv',
    'produccin-de-pozos-de-gas-y-petrleo-2021.csv',
    'produccin-de-pozos-de-gas-y-petrleo-2022.csv',
    'produccin-de-pozos-de-gas-y-petrleo-2023.csv'
]

# Create an empty list to store DataFrames
dfs = []

# Loop through each CSV file and read into a DataFrame
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    dfs.append(df)

# Concatenate the DataFrames vertically (along rows)
concatenated_df = pd.concat(dfs, ignore_index=True)

# # Save the concatenated DataFrame to a new CSV file
# concatenated_df.to_csv('C:/Users/Morris/Documents/produccin-de-pozos-de-gas-y-petrleo-2017-2023.csv', index=False)


```

```python
# # # Read the concatenated CSV file
# df = pd.read_csv('C:/Users/Morris/Documents/produccin-de-pozos-de-gas-y-petrleo-2017-2023.csv')

df = concatenated_df
```

```python
print(df.info())
```

```python
# List of columns to delete
columns_to_delete = ['idempresa', 'prod_agua', 'iny_agua', 'iny_gas', 'iny_co2', 'iny_otro', 'tef',
       'vida_util', 'observaciones', 'fechaingreso', 'rectificado', 'habilitado',
       'idusuario', 'sigla', 'formprod', 'formacion',
       'idareapermisoconcesion', 'areapermisoconcesion', 'idareayacimiento',
       'proyecto','sub_tipo_recurso']

# Delete the selected columns
df = df.drop(columns=columns_to_delete)
```

```python
# Print the column names
print("Column Names:")
print(df.columns)
```

```python
# #Categoricos
df.describe(include=object)
```

```python
# #Numericos
df.describe()
```

```python
## Valores unicos
# print("Afiliado:",len(df['Afiliado'].value_counts()))

for column in df:
    print('Column Name : ', column)
    print('Column freq : ', len(df[column].value_counts()))
    print()
```

```python
 #Frecuencias
for column in df:
  print('Column Name : ', column)
  print(pd.crosstab(df[column], "freq"))
  print()
```

```python
#Null values
df.isnull().sum()
```

```python
# Delete rows with prod_pet and prod_gas less than zero
df = df[(df['prod_pet'] >= 0) & (df['prod_gas'] >= 0)]
```
